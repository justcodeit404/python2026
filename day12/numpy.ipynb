```python
import pandas as pd
import numpy as np

us_file_path = r"D:\pyfile\USvideos.csv"

# 使用 pandas 读取 CSV（自动处理编码、混合类型）
df_us = pd.read_csv(us_file_path, encoding='utf-8')  # 显式指定 UTF-8

# 查看最后几列，确认评论列的名称（通常是 'comment_count'）
print(df_us.columns)

# 假设评论列叫 'comment_count'
t_us_comments = df_us['comment_count'].dropna().astype(int)  # 去除 NaN 并转为 int

print("最大评论数:", t_us_comments.max())
print("最小评论数:", t_us_comments.min())
```
